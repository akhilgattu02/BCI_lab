```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd

# Function to generate and plot a signal
def generate_and_plot_signal(signal, title):
    plt.figure(figsize=(10, 4))
    plt.plot(t, signal)
    plt.title(title)
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

# Function to listen to the signal
def listen_to_signal(signal, sampling_frequency):
    sd.play(signal, samplerate=sampling_frequency)
    sd.wait()
```

# Signal characteristics

-Base frequency: 10Hz

## Sine wave

```python
# Generate time values
t = np.linspace(0, 5, 5000)  # 5 seconds, 5000 samples
freq = 5 
sine_wave = np.sin(2 * np.pi * freq * t)
generate_and_plot_signal(sine_wave, 'Sine Wave')
listen_to_signal(sine_wave, 7000)
'''
# Sine wave
sine_wave = np.sin(2 * np.pi * t)
generate_and_plot_signal(sine_wave, 'Sine Wave')
listen_to_signal(sine_wave, 50000)
'''
```

## Linear chirp

```python
t = np.linspace(0, 5, 5000)  # 5 seconds, 5000 samples
freq = 5 * t
sine_wave = np.sin(2 * np.pi * freq * t)
generate_and_plot_signal(sine_wave, 'Sine Wave')
listen_to_signal(sine_wave, 7000)
```

## Quadratic chirp

```python
t = np.linspace(0, 5, 5000)  # 5 seconds, 5000 samples
freq = 5 * (t ** 2)
sine_wave = np.sin(2 * np.pi * freq * t)
generate_and_plot_signal(sine_wave, 'Sine Wave')
listen_to_signal(sine_wave, 7000)
```

## Logarithmic chirp

```python
t = np.linspace(0, 5, 5000)  # 5 seconds, 5000 samples
freq = 5 * (np.log(t))
sine_wave = np.sin(2 * np.pi * freq * t)
generate_and_plot_signal(sine_wave, 'Sine Wave')
listen_to_signal(sine_wave, 7000)
```
